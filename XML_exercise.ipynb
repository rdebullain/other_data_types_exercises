Using data from [**mondial database**](https://drive.google.com/file/d/14lFT4nWHgwN36ij4XZh6OUuup-K9qLgR/view?usp=sharing) find the answers to following questions:

2. 10 cities with the largest population.
3. name and country of a) longest river, b) largest lake and c) airport at highest elevation.

```python
import xml.etree.ElementTree as ET
```

```python
# Load and parse the XML file
tree = ET.parse('mondial.xml')
```

```python
# Get the root element
root = tree.getroot()
```

```python
# Iterate over the XML attributes
for child in root:
    print(child.tag, child.attrib)
```

## 1. 10 countries with the lowest infant mortality rates.

```python
# Initialize a list to store the data
data = []

# Iterate over the XML elements
for country in root.findall('.//country'):
    # Extract the country name
    country_name = country.find('name').text

    #Extract the infant mortality rate
    infant_mortality = country.find('infant_mortality')
    if infant_mortality is not None:
        infant_mortality = float(infant_mortality.text)
    else:
        continue # Skip this country if infant mortality is not available

    # Add the data to the list
    data.append((country_name, infant_mortality))

# Sort the list by infact mortality rate in ascending order
data.sort(key=lambda x: x[1])

# Select the first 10 elements
top_10 = data[:10]

# Print the results
for country, rate in top_10:
    print(f'{country}: {rate}')
    
```
